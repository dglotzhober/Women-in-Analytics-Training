Helper script to clean & merge OECD/IMF data and write out result, from and to `data/`.

```python
import pandas as pd

YVAR = "SW_LIFS"

oecd_stat = pd.read_csv("data/BLI_18122020161838856.csv", thousands=",")
imf_stat = pd.read_csv("data/WEO_Data.csv", thousands=",",)
imf_stat = imf_stat.pivot(index="Country", columns="Subject Descriptor", values="2019")
imf_stat.columns = ["GDPC", "UNEMP"]
oecd_stat = oecd_stat[oecd_stat["Inequality"] == "Total"]
oecd_stat = oecd_stat.pivot(index="Country", columns="INDICATOR", values="Value")
oecd_stat.drop("OECD - Total", inplace=True)
df = oecd_stat.merge(imf_stat, left_index=True, right_index=True)
df.dropna(subset=[YVAR], inplace=True)
df = df.fillna(df.median(axis=0))
df.to_csv("data/country_satis.csv", encoding="utf-8")
```
