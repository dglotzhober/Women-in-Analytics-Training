```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
```

The method from statsmodels.api is called OLS() and it takes two arguments, y and x. After initiating the OLS model you can call the .fit() method to have it fit on the data. Read more in the documentation located here: https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.OLS.html



```python
data = pd.DataFrame({'hours_of_sunshine' : np.arange(0,11), 'popsicles_sold' : [2, 4, 7, 6, 8, 13, 12, 18, 22, 22, 19]})
```

```python
data
```

```python
x = data['hours_of_sunshine']
y = data['popsicles_sold']

plt.scatter(x, y)
```

```python
sns.regplot(x, y)
```
